```python
import pandas as pd
```

```python
train = '/kaggle/input/home-data-for-ml-course/train.csv'
train = pd.read_csv(train) 
train.describe()
```

```python
test = '/kaggle/input/home-data-for-ml-course/test.csv'
test = pd.read_csv(test)
test.describe()
```

```python
y_train = train.SalePrice
```

```python
train.head()
```

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
train_data = train[features]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae_rf(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)

def get_mae_dt(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python

train_X, val_X, train_y, val_y = train_test_split(train_data, y_train, random_state=1)
```

```python
for max_leaf_nodes in [5, 50, 500, 600]:
    my_mae = get_mae_rf(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes for Random Forest: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))

print('-'*100)
for max_leaf_nodes in [10,100,1000]:
    my_mae = get_mae_dt(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes for Decision Tree: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
rf_model = RandomForestRegressor(random_state=1)
```

```python
rf_model.fit(train_X, train_y)
```

```python
rf_val_predictions = rf_model.predict(val_X)
```

```python
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)
```

```python

print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```

```python
rf_model_on_full_data = RandomForestRegressor(random_state=1)
rf_model_on_full_data.fit(train_data,y_train)
```
